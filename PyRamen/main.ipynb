```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
#menu = []
#sales = []

def csv_list(csvFile):
    outputList = list()
    reader = csv.reader(open(csvFile))
    next(reader)
    for row in reader:
        outputList.append(row)            
    return outputList  

# @TODO: Read in the menu data into the menu list
# @TODO: Read in the sales data into the sales list
def build_report(menu, sales):
        
    # @TODO: Initialize dict object to hold our key-value pairs of items and metrics
    report = {}

    # @TODO: Loop over every row in the sales list object
    for irow in sales:

        # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
        # @TODO: Initialize sales data variables
        if irow != []:
            quantity = irow[3]
            sales_item = irow[4]
        else:
            break

        # @TODO:
        # If the item value not in the report, add it as a new entry with initialized metrics
        # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
        if sales_item not in report:
            report[sales_item] = {'01-count': 0, '02-revenue': 0, '03-cost':0, '04-profit':0}
            
        # @TODO: For every row in our sales data, loop over the menu records to determine a match
        for menu_item in menu:
           
            # Item,Category,Description,Price,Cost
            # @TODO: Initialize menu data variables
            m_item_name = menu_item[0]
            m_item_price = menu_item[3]
            m_item_cost = menu_item[4]

            # @TODO: Calculate profit of each item in the menu data
            profit = float(m_item_price) - float(m_item_cost)

            # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item
            if sales_item == m_item_name:

                # @TODO: Print out matching menu data
                report[sales_item]['01-count'] += int(quantity)
                report[sales_item]['02-revenue'] += float(m_item_price) * int(quantity)
                report[sales_item]['03-cost'] += float(m_item_cost) * int(quantity)
                report[sales_item]['04-profit'] += profit * int(quantity)    
          
    return report
            # @TODO: Cumulatively add up the metrics for each item key
       # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
    # @TODO: Increment the row counter by 1


# @TODO: Write out report to a text file (won't appear on the command line output)

menu_data = Path(menu_filepath)
sales_data = Path(sales_filepath)
report_path = Path("Resources/report.txt")

# Initialize list objects to hold our menu and sales data
menu = csv_list(menu_data)
sales = csv_list(sales_data)

# Generate the report
report = build_report(menu, sales)

# Write the report to the report file
with open(report_path, 'w') as report_file:
    report_file.write('## PyRamen Report\n')
    report_file.write('\n|Item|Count|Revenue|Cost|Profit|\n')
    report_file.write('|---------------------------------------|\n')
    for item in report:
        cur = report[item]
        report_file.write(f'|{item.title()}|{cur["01-count"]}|{cur["02-revenue"]}|{cur["03-cost"]}|{cur["04-profit"]}|\n')


# Write the report as a list in markdown format
with open(report_path, 'a') as report_file:
   
    for item in report:
        cur_item = report[item]
        report_file.write(f'\n## -> {item.title()}\n')
        report_file.write(f'\n01 - Count: {cur_item["01-count"]}\n')
        report_file.write(f'02 - Revenue: {cur_item["02-revenue"]}\n')
        report_file.write(f'03 - Cost: {cur_item["03-cost"]}\n')
        report_file.write(f'04 - Profit: {cur_item["04-profit"]}\n')
        
    # @TODO: Print total number of records in sales data           
    with open(sales_data, 'r') as sales_file:
        csv_reader = csv.reader(sales_file, delimiter=',')
        rec_count = 0
        for row in sales_file:
            rec_count += 1
        
        report_file.write('\n-----------------------------------------\n')
        report_file.write(f'  Total Number of Sales Records: {rec_count}\n')
        report_file.write('-----------------------------------------\n')
```
