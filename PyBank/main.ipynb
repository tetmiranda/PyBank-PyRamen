```python
# 'os' is a built in module in python  
# a useful functions that can be used to list directory contents and filter the results
import os
import csv

# Set the path to collect data from the Resources folder
budget_data_csv = os.path.join("Resources", "budget_data.csv")

# Initialise the variables
months = []
chgs_profit_loss = []
count_mos = 0

curr_mos_profit_loss = 0
prev_mos_profit_loss = 0
profit_loss_change = 0

net_profit_loss = 0

sum_profit_loss = 0
avg_profit_loss = 0
highest_change = 0
lowest_change = 0

# open budget data CSV using the set path budget_data_csv
with open(budget_data_csv, newline="") as csvf:
    
    csv_reader = csv.reader(csvf, delimiter=",")
    
    # Read the header row
    csv_header = next(csv_reader) 
    
    # Read through each of data after the header
    for row in csv_reader:
        
        # Count the months
        count_mos += 1
        
        # Net total amount of "Profit/Losses" over the entire period
        curr_mos_profit_loss = int(row[1])
        net_profit_loss = net_profit_loss + curr_mos_profit_loss
        
        if (count_mos == 1):
            # Take the current month to be equal to the previous month
            prev_mos_profit_loss = curr_mos_profit_loss

        else:
            
            # Calculate change in profit loss previous date profit - latest or current date profit e.g. B2-B3 when in excel
            profit_loss_change = curr_mos_profit_loss - prev_mos_profit_loss

            # Append each month to the months[]
            months.append(row[0])
 
            # Append each profit_loss_change to the profit loss changes
            chgs_profit_loss.append(profit_loss_change)
            
            # Make the current_month_loss to be previous_month_profit_loss for the next loop
            prev_mos_profit_loss = curr_mos_profit_loss
    
    #sum and average of the changes in "Profit/Losses" over the entire period
    sum_profit_loss = sum(chgs_profit_loss)
    avg_profit_loss = round(sum_profit_loss/(count_mos - 1), 2)

    # highest and lowest changes in "Profit/Losses" over the entire period
    highest_change = max(chgs_profit_loss)
    lowest_change = min(chgs_profit_loss)

    # Locate the index value of highest and lowest changes in "Profit/Losses" over the entire period
    highest_month_index = chgs_profit_loss.index(highest_change)
    lowest_month_index = chgs_profit_loss.index(lowest_change)

    # Assign best and worst month
    best_month = months[highest_month_index]
    worst_month = months[lowest_month_index]

# -->>  Print the analysis to the terminal
print("Financial Analysis")
print("----------------------------")
print(f"Total Months:  {count_mos}")
print(f"Total:  ${net_profit_loss}")
print(f"Average Change:  ${avg_profit_loss}")
print(f"Greatest Increase in Profits:  {best_month} (${highest_change})")
print(f"Greatest Decrease in Losses:  {worst_month} (${lowest_change})")

# Set the path to store data into the new text file from the Resources folder
fin_analysis_txt = os.path.join("Resources", "financial_analysis.txt")

with open(fin_analysis_txt, 'w') as text:
    text.write("----------------------------------------------------------\n")
    text.write("  Financial Analysis"+ "\n")
    text.write("----------------------------------------------------------\n\n")
    text.write("  Total Months: " + str(count_mos) + "\n")
    text.write("  Total Profits: " + "$" + str(net_profit_loss) +"\n")
    text.write("  Average Change: " + '$' + str(int(avg_profit_loss)) + "\n")
    text.write("  Greatest Increase in Profits: " + str(best_month) + " ($" + str(highest_change) + ")\n")
    text.write("  Greatest Decrease in Profits: " + str(worst_month) + " ($" + str(lowest_change) + ")\n")
    text.write("----------------------------------------------------------\n")
```
